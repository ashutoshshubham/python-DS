```python
weatherdf = pd.read_csv('weather.csv',parse_dates = ['Date.Full'],dayfirst = True)
weatherdf.head()
```

```python
print(weatherdf.columns.tolist())
```

```python
weatherdf.columns = ['precipitation','date','month','week','year','city','code','location','state','avg_temp','max_temp','min_temp','wind_direction','wind_speed']
```

```python
weatherdf.head()
```

```python
weatherdf.drop(columns = ['code','location'],inplace = True)   #test run without inplace=True, then add and run again
```
